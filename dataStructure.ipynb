```python
#list
"""------------------------------------------------------------------------------"""
#2배속
myList = [5, 'banana', 'cherry',"apple"]
print(len(myList))
#remove
item = myList.remove(5)
print(myList,item)
#sort
item = myList.sort()
print(myList,item)

#[0] * 5
numList = [0] * 5
rangeList = list(range(1,5))
print(numList+rangeList)

#list slicing
#number in end doesn't include to new lists.
#[start,end,step]
print(rangeList[2:3])
print(rangeList[0:5:2])

#copy
list_orgin = [1,2,3,4,5]
#얕은 복사
list_copy = list_orgin
list_orgin[3]=61

print(list_copy)
#deep copy
list_copy2=list_orgin[:]
list_orgin[2]=61

print(list_orgin)
print(list_copy2)


#list comprehension and if
newList = [1,2,3,4,5,6,7]
b= [i*i for i in newList if i%2==0]
print(b)

```

```python
#tuple
# 변하지 않는, 중복을 허용하는 구조
import sys
import timeit
myTuple = tuple(["MAX", 28,"123","Boston"])
print(myTuple)
for i in myTuple:
    print(i)
if "MAX" in myTuple:
    print("yes")

appleTuple = ("a","p","p","l","e")
print(appleTuple.count("p"))
#첫번째만 리턴
print(appleTuple.index("p"))

b= appleTuple[::-1]
print(b)
#pack, unpack
a1,a2,a3,a4,a5= appleTuple
print(a1,a2,a3,a4,a5)
b1,*b2=appleTuple
print(b1,b2)

#getsizeof

list1=[0,1,2,3,"spstar",True]
tuple1= (0,1,2,3,"spstar",True)
print(sys.getsizeof(list1),list1)
print(sys.getsizeof(tuple1),tuple1)


#timeit method => 특정 횟수만큼 명령어를 실행시키고, 그 효율성을 본다.
print(timeit.timeit(stmt="[0,1,2,3,4,5,6,7]",number=1000000))
print(timeit.timeit(stmt="(0,1,2,3,4,5,6,7)",number=1000000))
```

```python
#dict key-value

my_dict = {"name" : "MAX", "age" : 28, "city": "Anyang"}
print(my_dict)

#del elements
# del my_dict["name"]
my_dict.pop('name')
print(my_dict)

if "name" not in my_dict:
    my_dict['name'] ='gichun'
try :
    print(my_dict['name1'])
except :
    print("Error")


#iter key, value, items
for key in my_dict:
    print(key)
for key in my_dict.keys():
    print(key)
for value in my_dict.values():
    print(value)
for key,value in my_dict.items():
    print(key,value)

#얕은 복사
copy_dict =my_dict
copy_dict["copy"]=1
print(my_dict)

#deep copy
copy_dict2 = my_dict.copy()
copy_dict2["copy"]=2
print(my_dict)

#update
copy_dict2["gender"]= 'male'
copy_dict.update(copy_dict2)
print(copy_dict)

#list cannot be keys
```

```python
#sets
myset = {1,2,3,2,3}
print(myset)
strSet = set("hello")
print(strSet)

#add ,remove
strSet.add("q")
#raise error if set doesn't have an element
strSet.remove("o")
#no error 
strSet.discard("o")
#pop
print(strSet)
print(strSet.pop())

#union, intersect, etc..
ToNine = list(range(10))
odds = set(ToNine[1:10:2])
evens = set(ToNine[0:10:2])
primes = {2,3,5,7,}

print(odds.union(evens))
print(odds.intersection(evens))
print(odds.difference(primes))
print(odds -  primes)

#A u B - A n B
print(odds.symmetric_difference(primes))
# odds.update(evens)
print(odds)

#부분집합 확인

odds2 = {3,5}
print(odds2.issubset(odds))

#서로소 확인
print(evens.isdisjoint(odds))

#frozenset ->cannot change
a = frozenset([1,2,3,4,5])
```
